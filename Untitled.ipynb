```python
# Function to load cached data
def load_cached_data(cache_path, description="data"):
    """Load cached data from pickle file"""
    try:
        if os.path.exists(cache_path):
            print(f"Loading {description} from cache: {cache_path}")
            with open(cache_path, 'rb') as f:
                result = pickle.load(f)
                print(f"Successfully loaded {description}")
                return result
        else:
            print(f"Cache file not found: {cache_path}")
            return None
    except Exception as e:
        print(f"Error loading {description} from cache: {e}")
        traceback.print_exc()
        return None

# Load GRN tables data if available
def load_grn_tables_data():
    grn_tables_pkl = os.path.join(grn_tables_dir, 'grn_tables_data.pkl')
    if os.path.exists(grn_tables_pkl):
        try:
            with open(grn_tables_pkl, 'rb') as f:
                grn_data = pickle.load(f)
                print(f"Successfully loaded GRN tables data")
                return grn_data
        except Exception as e:
            print(f"Error loading GRN tables data: {e}")
            traceback.print_exc()
    else:
        print(f"GRN tables data file not found: {grn_tables_pkl}")
    return None

# Load main workflow data
chain_id = 'A'

# Try to load all cached data from the workflow
cache_files = {
    "raw_structures": (f"raw_structures_{chain_id}.pkl", "raw structures"),
    "processed_structures": (f"processed_structures_{chain_id}.pkl", "processed structures"),
    "structure_errors": (f"structure_errors_{chain_id}.pkl", "structure errors"),
    "helix_annotations": (f"helix_annotations_{chain_id}.pkl", "helix annotations"),
    "structure_comparison": (f"structure_comparison_{chain_id}.pkl", "structure comparison"),
    "grn_assignment": (f"grn_assignment_{chain_id}.pkl", "GRN assignment")
}

for data_type, (file_name, description) in cache_files.items():
    cache_path = os.path.join(cache_dir, file_name)
    component_data = load_cached_data(cache_path, description)
    if component_data:
        # If the component data is a dictionary, update the main data dictionary
        if isinstance(component_data, dict):
            data.update(component_data)
        # Otherwise store it directly
        else:
            data[data_type] = component_data

# Additionally, load the GRN tables data
grn_data = load_grn_tables_data()
if grn_data:
    data.update(grn_data)
    
# Print confirmation of data loading
print(f"\nLoaded data with {len(data.keys())} top-level keys")
if 'processed_structures' in data:
    print(f"Found {len(data['processed_structures'])} processed structures")
```

```python
# Check available components and datasets in the loaded data
def print_data_summary(data):
    """Print summary of loaded data"""
    print("\n=== DATA SUMMARY ===")
    
    # Print top-level keys
    print("\nTop-level components:")
    for key in sorted(data.keys()):
        print(f"- {key}")
    
    # Print dataset information if available
    if 'datasets' in data:
        print("\nDatasets:")
        for dataset_id, dataset in data.get('datasets', {}).items():
            content = dataset.get('content', [])
            count = len(content) if isinstance(content, list) else 'Unknown'
            print(f"- {dataset_id}: {count} structures")
    
    # Print processor information if available
    if 'processor' in data:
        print("\nProcessor:")
        processor = data['processor']
        processor_type = type(processor).__name__
        print(f"- Type: {processor_type}")
        
    # Print structure counts if available
    if 'processed_structures' in data:
        processed = data.get('processed_structures', {})
        print(f"\nProcessed structures: {len(processed)}")
        
    # Print mapping info if available
    if 'structure_mapping' in data:
        mapping = data.get('structure_mapping', {})
        print(f"\nStructure mapping pairs: {len(mapping)}")
        
    # Print GRN info if available
    if 'msa_table' in data:
        print("\nMSA table available")
        msa = data.get('msa_table', pd.DataFrame())
        if isinstance(msa, pd.DataFrame):
            print(f"- Shape: {msa.shape}")
    
    # Print tables info
    for table_name in ["ca_distance_table_grn", "distance_table_grn", "ca_residue_table_grn", 
                       "residue_table_grn", "column_mapping"]:
        if table_name in data:
            table = data.get(table_name)
            if isinstance(table, pd.DataFrame):
                print(f"\n{table_name} table available")
                print(f"- Shape: {table.shape}")

# Print summary of loaded data
if data:
    print_data_summary(data)
```

```python
# Retinal-related data inspection and debugging
import re
import pandas as pd # Assuming pandas is imported, as it's used in your function

# --- Your existing code from the question ---
# (Make sure 'data' is loaded before this script runs, e.g., from a pickle file or previous cell)
# Example:
# import pickle
# with open(os.path.join(cache_dir, 'data_dictionary.pkl'), 'rb') as f:
#     data = pickle.load(f)
# Or if 'data' is defined globally in your notebook already, that's fine.


# First check all unique residue types in the processed structures
def check_residue_types(data_dict):
    """
    Check all unique residue types in the loaded structures
    
    Args:
        data_dict (dict): Data dictionary from the workflow
        
    Returns:
        dict: Dictionary of unique residue types by structure
    """
    residue_types = {}
    
    if 'processed_structures' not in data_dict:
        print("No processed structures found in data.")
        return residue_types
    
    processed_structures = data_dict['processed_structures']
    
    for pdb_id, struct_data in processed_structures.items():
        # Check main structure DataFrame
        if 'df' in struct_data and isinstance(struct_data['df'], pd.DataFrame):
            df = struct_data['df']
            if 'res_name3l' in df.columns:
                unique_residues = sorted(df['res_name3l'].unique().tolist())
                residue_types[pdb_id] = unique_residues
            else:
                # Handle case where 'res_name3l' column might be missing for a structure
                print(f"Warning: 'res_name3l' column not found in DataFrame for PDB ID: {pdb_id}")
                residue_types[pdb_id] = [] # Assign empty list if column is missing
        else:
            # Handle case where 'df' might be missing or not a DataFrame
            print(f"Warning: DataFrame ('df') not found or not a DataFrame for PDB ID: {pdb_id}")
            residue_types[pdb_id] = [] # Assign empty list
    
    return residue_types

# !!! IMPORTANT: Ensure 'data' is defined before this line !!!
# For example, if it's loaded from a file:
# try:
#     with open(os.path.join(cache_dir, 'your_data_file.pkl'), 'rb') as f: # Replace with your actual data file
#         data = pickle.load(f)
# except FileNotFoundError:
#     print("Error: Data file not found. Please ensure 'data' is loaded.")
#     data = {'processed_structures': {}} # Provide a default empty structure to avoid further errors
# except Exception as e:
#     print(f"Error loading data: {e}")
#     data = {'processed_structures': {}}


# Get unique residue types by structure
# Ensure 'data' is loaded and valid before calling this
if 'data' in locals() and isinstance(data, dict):
    residue_types_by_structure = check_residue_types(data)
else:
    print("Error: 'data' variable is not defined or not a dictionary. Cannot perform residue type analysis.")
    residue_types_by_structure = {} # Initialize to empty to prevent downstream errors

# Analyze all observed residue types
all_residue_types = set()
for pdb_id, residues in residue_types_by_structure.items():
    all_residue_types.update(residues)

print("\n=== RETINAL AND RESIDUE TYPE ANALYSIS ===")
if residue_types_by_structure: # Only print if there's something to analyze
    print(f"Analyzed {len(residue_types_by_structure)} structures")
    print(f"Found {len(all_residue_types)} unique residue types across all structures")

    # Count structures with each residue type
    residue_counts = {}
    for residue_type in all_residue_types:
        count = sum(1 for residues in residue_types_by_structure.values() if residue_type in residues)
        residue_counts[residue_type] = count

    # Print most common residue types
    print("\nMost common residue types:")
    for residue, count in sorted(residue_counts.items(), key=lambda x: x[1], reverse=True)[:20]:
        print(f"  {residue}: {count} structures")

    # Check specifically for retinal-related residues
    retinal_residues = ['RET', 'LIG', 'LYR'] # LIG can sometimes be retinal, LYR is Lysine-Retinal
    print("\nRetinal-related residue types:")
    for residue in retinal_residues:
        count = residue_counts.get(residue, 0)
        structures = [pdb_id for pdb_id, residues_list in residue_types_by_structure.items() if residue in residues_list]
        print(f"  {residue}: {count} structures")
        if 0 < count <= 10: # Only print list if few structures, to avoid long outputs
            print(f"    Found in: {', '.join(structures)}")

    # Check for LYR residues specifically
    lyr_structures = [pdb_id for pdb_id, residues in residue_types_by_structure.items() if 'LYR' in residues]
    if lyr_structures:
        print(f"\nFound LYR residues in {len(lyr_structures)} structures: {lyr_structures}")
        
        # Examine LYR residues in detail (if 'data' is available)
        if 'data' in locals() and 'processed_structures' in data:
            for pdb_id in lyr_structures:
                if pdb_id in data['processed_structures']:
                    struct = data['processed_structures'][pdb_id]
                    if 'df' in struct and isinstance(struct['df'], pd.DataFrame):
                        df = struct['df']
                        lyr_mask = df['res_name3l'] == 'LYR'
                        lyr_data = df[lyr_mask]
                        
                        if not lyr_data.empty:
                            print(f"\nLYR residue details for {pdb_id}:")
                            print(f"  Atom count: {len(lyr_data)}")
                            
                            if 'auth_seq_id' in lyr_data.columns:
                                lyr_residue_ids = lyr_data['auth_seq_id'].unique()
                                print(f"  Residue positions: {lyr_residue_ids}")
                            
                            if 'res_atom_name' in lyr_data.columns:
                                atom_names = sorted(lyr_data['res_atom_name'].unique())
                                print(f"  Atom names: {atom_names}")
    else:
        print("\nNo LYR residues found in loaded data.")

    # Check other unusual retinal designations
    other_retinal_candidates = ['RAL', 'RTL', 'RTN', 'RNL', 'RET-A', 'A-RET']
    unusual_structures = []

    for residue_candidate in other_retinal_candidates: # Renamed 'residue' to 'residue_candidate' to avoid conflict
        structures = [pdb_id for pdb_id, residues_list in residue_types_by_structure.items()
                     if any(r.startswith(residue_candidate) for r in residues_list)]
        if structures:
            unusual_structures.extend(structures) # Add to list of structures with unusual retinal
            print(f"\nFound potential retinal residue '{residue_candidate}' in {len(structures)} structures: {structures}")

    if unusual_structures: # Check if the list is not empty
        print(f"\nFound {len(set(unusual_structures))} unique structures with unusual retinal designations")
    else:
        print("\nNo unusual retinal designations found among candidates.")

else: # If residue_types_by_structure is empty
    print("No residue types to analyze.")


# --- NEW CODE TO PRINT STRUCTURES WITHOUT 'RET' ---
print("\n=== STRUCTURES WITHOUT 'RET' RESIDUE ===")
if not residue_types_by_structure:
    print("No structures were analyzed, so cannot determine which are without 'RET'.")
else:
    structures_without_ret = []
    for pdb_id, residues in residue_types_by_structure.items():
        if 'RET' not in residues:
            structures_without_ret.append(pdb_id)

    if structures_without_ret:
        print(f"Found {len(structures_without_ret)} structures without 'RET':")
        # Sort for consistent output, especially if there are many
        for pdb_id in sorted(structures_without_ret):
            print(f"  {pdb_id}")
    else:
        # This case means all analyzed PDBs had 'RET' or residue_types_by_structure was empty initially.
        # The outer 'if not residue_types_by_structure' handles the empty case more directly.
        print("All analyzed structures contain 'RET'.")

# --- End of new code ---
```

```python
import pandas as pd
import re
import numpy as np # Not strictly needed here but good for general DS
import os

# --- Re-use or adapt sorting logic from your msa_grn.py or similar ---
# This is a simplified version for demonstration. You should use your robust one.
def grn_to_sortable_value(label):
    """
    Converts a GRN label to a sortable tuple.
    Handles: n.X, H.X (1-7), LH.X (loop between L and H), c.X, X.X (unclassified)
    Corrected to handle multi-digit positions properly.
    """
    if not isinstance(label, str):
        return (99, 999, 999) # Sort non-strings last

    # N-terminal: n.XX
    match_n_term = re.fullmatch(r'n\.(\d+)', label)
    if match_n_term:
        # N-terminal regions, sorted reverse numerically (n.1 is "latest" before TM1)
        return (0, -int(match_n_term.group(1)), 0)

    # TM helices: H.YY (e.g., 1.50, 7.43) - Helix 1-7
    match_tm = re.fullmatch(r'([1-7])\.(\d+)', label)
    if match_tm:
        helix, pos = map(int, match_tm.groups())
        return (helix, pos, 0)

    # Loop regions: HL.DDD (e.g., 12.001, 67.010) - H=prev helix, L=next helix
    match_loop = re.fullmatch(r'([1-7])([1-7])\.(\d+)', label)
    if match_loop:
        h1, h2, pos = map(int, match_loop.groups())
        # Sort by the first helix number of the pair, then by position
        return (min(int(h1), int(h2)) + 0.5, pos, 0) # Place loops between helices

    # C-terminal: c.XX
    match_c_term = re.fullmatch(r'c\.(\d+)', label)
    if match_c_term:
        return (8, int(match_c_term.group(1)), 0) # Primary sort key after TM7

    # Fallback for other formats like X.YY or L.YY (generic loop/unclassified)
    # This part needs to be robust if you have many 'other' formats.
    match_other_l = re.fullmatch(r'L\.(\d+)', label)
    if match_other_l: # Generic Loop
        return (7.6, int(match_other_l.group(1)), 0) # After last loop, before C-term
        
    match_other_x = re.fullmatch(r'X\.(\d+)', label)
    if match_other_x: # Unclassified
         return (9, int(match_other_x.group(1)), 0) # After C-term
    
    # For anything else, try to give it a very high sort order or parse more specifically
    # A simple hash can be used for unrecognised patterns, but explicit parsing is better.
    return (99, hash(label), 0)


def sort_grn_column_names(df_columns: list) -> list:
    """Sorts a list of GRN column labels."""
    return sorted(df_columns, key=grn_to_sortable_value)


def parse_grn_components(grn_label_str: str):
    """
    Parses a GRN string into components for easier processing.
    Returns: (type, part1, part2, original_label)
    type: 'tm', 'n_term', 'c_term', 'loop', 'other_L', 'other_X', 'unparsed'
    part1: helix_num (tm), 'n' (n_term), 'c' (c_term), loop_pair_str ("12"), 'L', 'X'
    part2: position_num (tm, n_term, c_term, loop, L, X)
    """
    if not isinstance(grn_label_str, str):
        return ('unparsed', grn_label_str, None, grn_label_str)

    m = re.fullmatch(r'([1-7])\.(\d+)', grn_label_str)
    if m: return ('tm', int(m.group(1)), int(m.group(2)), grn_label_str)
    
    m = re.fullmatch(r'n\.(\d+)', grn_label_str)
    if m: return ('n_term', 'n', int(m.group(1)), grn_label_str)
    
    m = re.fullmatch(r'c\.(\d+)', grn_label_str)
    if m: return ('c_term', 'c', int(m.group(1)), grn_label_str)
    
    m = re.fullmatch(r'([1-7])([1-7])\.(\d{1,3})', grn_label_str)
    if m:
        h1, h2, pos = m.groups()
        pair = "".join(sorted((h1, h2))) # e.g. "12"
        return ('loop', pair, int(pos), grn_label_str)
        
    m = re.fullmatch(r'L\.(\d+)', grn_label_str)
    if m: return ('other_L', 'L', int(m.group(1)), grn_label_str)

    m = re.fullmatch(r'X\.(\d+)', grn_label_str)
    if m: return ('other_X', 'X', int(m.group(1)), grn_label_str)
    
    return ('unparsed', grn_label_str, None, grn_label_str)


def postprocess_grn_msa_table(input_csv_path: str, output_csv_path: str,
                              check_lys_7_50: bool = True):
    """
    Loads a GRN-labeled MSA table, ensures all canonical GRN positions within
    the observed range for each segment are present and consecutive, and saves the fixed table.

    Args:
        input_csv_path: Path to the input CSV file (GRN-labeled MSA).
        output_csv_path: Path to save the fixed CSV file.
        check_lys_7_50: If True, checks if position 7.50 is predominantly Lysine.
    """
    try:
        msa_df = pd.read_csv(input_csv_path, index_col=0)
        # Ensure all data is string to handle mixed types from read_csv if some cols are all numeric
        msa_df = msa_df.astype(str)
    except FileNotFoundError:
        print(f"Error: Input file not found at {input_csv_path}")
        return
    except Exception as e:
        print(f"Error loading CSV: {e}")
        return

    print(f"Loaded MSA table with {msa_df.shape[0]} sequences and {msa_df.shape[1]} GRN positions.")

    # --- 1. Parse all existing GRN columns and group them ---
    # Structure: parsed_columns[col_type][segment_id] = [pos1, pos2, ...]
    # e.g., parsed_columns['tm'][1] = [48, 50, 51]
    # e.g., parsed_columns['loop']['12'] = [1, 3]
    parsed_columns = {}
    unparsed_labels_set = set()

    for col_label in msa_df.columns:
        col_type, p1_id, p2_pos, _ = parse_grn_components(col_label)

        if col_type == 'unparsed' or p2_pos is None : # p2_pos must be a number
            unparsed_labels_set.add(col_label)
            continue

        if col_type not in parsed_columns:
            parsed_columns[col_type] = {}
        if p1_id not in parsed_columns[col_type]:
            parsed_columns[col_type][p1_id] = []
        parsed_columns[col_type][p1_id].append(p2_pos)

    # --- 2. Generate the full list of expected GRN columns based on observed ranges ---
    all_expected_grn_labels = set()

    for col_type, segments in parsed_columns.items():
        for segment_id, positions in segments.items():
            if not positions: continue
            min_pos, max_pos = min(positions), max(positions)
            for pos_val in range(min_pos, max_pos + 1):
                if col_type == 'tm':
                    all_expected_grn_labels.add(f"{segment_id}.{pos_val}")
                elif col_type == 'n_term': # segment_id is 'n'
                    all_expected_grn_labels.add(f"n.{pos_val}")
                elif col_type == 'c_term': # segment_id is 'c'
                    all_expected_grn_labels.add(f"c.{pos_val}")
                elif col_type == 'loop': # segment_id is "12", "23" etc.
                    all_expected_grn_labels.add(f"{segment_id}.{pos_val:03d}") # Ensure 3-digit padding for loops
                elif col_type == 'other_L': # segment_id is 'L'
                    all_expected_grn_labels.add(f"L.{pos_val}")
                elif col_type == 'other_X': # segment_id is 'X'
                    all_expected_grn_labels.add(f"X.{pos_val}")
    
    # Add any unparsed labels back
    all_expected_grn_labels.update(unparsed_labels_set)

    # --- 3. Sort all generated GRN labels ---
    sorted_expected_grn_labels = sort_grn_column_names(list(all_expected_grn_labels))
    print(f"Generated {len(sorted_expected_grn_labels)} expected GRN columns after filling gaps.")

    # --- 4. Create the new DataFrame ---
    fixed_msa_df = pd.DataFrame(index=msa_df.index)
    num_inserted_cols = 0
    for grn_label in sorted_expected_grn_labels:
        if grn_label in msa_df.columns:
            fixed_msa_df[grn_label] = msa_df[grn_label]
        else:
            fixed_msa_df[grn_label] = '-' # Insert a column of gaps
            num_inserted_cols += 1
    
    print(f"Inserted {num_inserted_cols} new columns (gaps) to ensure consecutiveness.")
    print(f"Final fixed MSA table has {fixed_msa_df.shape[0]} sequences and {fixed_msa_df.shape[1]} GRN positions.")

    # --- 5. Optional: Check for Lysine at 7.50 ---
    if check_lys_7_50 and '7.50' in fixed_msa_df.columns:
        lys_count = 0
        total_non_gap = 0
        warnings_7_50 = []
        for protein_id, residue_val_cell in fixed_msa_df['7.50'].items():
            residue_val = str(residue_val_cell) # Ensure it's a string
            if residue_val != '-':
                total_non_gap += 1
                # residue_val is like 'K282' or 'A280' or just 'K' or 'A'
                aa = residue_val[0] 
                if aa.isalpha(): # Make sure the first char is an AA
                    if aa.upper() == 'K':
                        lys_count += 1
                    else:
                        warnings_7_50.append(f"  - Protein {protein_id}: 7.50 is {residue_val} (not K)")
                else:
                    warnings_7_50.append(f"  - Protein {protein_id}: 7.50 is {residue_val} (unexpected format)")

        if total_non_gap > 0:
            lys_percentage = (lys_count / total_non_gap) * 100
            print(f"Check for Lysine at 7.50: {lys_count}/{total_non_gap} ({lys_percentage:.1f}%) are Lysine.")
            if warnings_7_50:
                print("Warnings for 7.50 not being Lysine or having unexpected format:")
                for warning in warnings_7_50[:5]: # Print first 5 warnings
                    print(warning)
                if len(warnings_7_50) > 5:
                    print(f"  ... and {len(warnings_7_50)-5} more.")
        else:
            print("Check for Lysine at 7.50: No non-gap residues found at this position or column '7.50' not present.")

    # --- 6. Save the fixed DataFrame ---
    try:
        fixed_msa_df.to_csv(output_csv_path)
        print(f"Successfully saved fixed MSA table to {output_csv_path}")
    except Exception as e:
        print(f"Error saving fixed MSA table: {e}")

    return fixed_msa_df
```

```python
# Analyze retinal properties in the loaded data
def analyze_retinal_properties(data, retinal_names=['RET', 'LIG']):
    """
    Analyze retinal properties in the loaded structures
    
    Args:
        data (dict): Data dictionary from the workflow
        retinal_names (list): List of possible retinal residue names to check
        
    Returns:
        dict: Dictionary of retinal properties by structure ID
    """
    retinal_properties = {}
    
    if 'processed_structures' not in data:
        print("No processed structures found in data.")
        return retinal_properties
    
    processed_structures = data['processed_structures']
    
    for pdb_id, struct_data in processed_structures.items():
        # Check if we have retinal data
        has_retinal = False
        retinal_atoms = []
        retinal_residues = set()
        retinal_types_found = set()
        
        # Check main structure DataFrame
        if 'df' in struct_data and isinstance(struct_data['df'], pd.DataFrame):
            df = struct_data['df']
            if 'res_name3l' in df.columns:
                # Check all possible retinal names
                for ret_name in retinal_names:
                    ret_mask = df['res_name3l'] == ret_name
                    if ret_mask.any():
                        retinal_df = df[ret_mask]
                        has_retinal = True
                        retinal_types_found.add(ret_name)
                        
                        if 'res_atom_name' in retinal_df.columns:
                            retinal_atoms.extend(retinal_df['res_atom_name'].tolist())
                        
                        # Get retinal residue IDs
                        if 'auth_seq_id' in retinal_df.columns:
                            retinal_residues.update(retinal_df['auth_seq_id'].unique())
        
        # Check dedicated retinal DataFrame
        if 'df_ret' in struct_data and isinstance(struct_data['df_ret'], pd.DataFrame):
            df_ret = struct_data['df_ret']
            if not df_ret.empty:
                has_retinal = True
                if 'res_name3l' in df_ret.columns:
                    retinal_types_found.update(df_ret['res_name3l'].unique())
                
                if 'res_atom_name' in df_ret.columns:
                    retinal_atoms.extend(df_ret['res_atom_name'].tolist())
                
                # Get retinal residue IDs
                if 'auth_seq_id' in df_ret.columns:
                    retinal_residues.update(df_ret['auth_seq_id'].unique())
        
        # Store properties
        retinal_properties[pdb_id] = {
            'has_retinal': has_retinal,
            'retinal_types': sorted(list(retinal_types_found)),
            'retinal_residues': sorted(list(retinal_residues)),
            'retinal_atom_count': len(retinal_atoms),
            'unique_retinal_atoms': sorted(list(set(retinal_atoms))) if retinal_atoms else []
        }
    
    return retinal_properties

# Analyze retinal residue atoms with expanded retinal names
retinal_names = ['RET', 'LIG', 'LYR', 'RAL', 'RTL', 'RTN', 'RNL']
retinal_properties = analyze_retinal_properties(data, retinal_names)

# Print summary of retinal properties
print("\n=== RETINAL PROPERTIES ===")
print(f"Analyzed retinal properties for {len(retinal_properties)} structures")

structures_with_retinal = [pdb_id for pdb_id, props in retinal_properties.items() if props['has_retinal']]
print(f"Found retinal in {len(structures_with_retinal)}/{len(retinal_properties)} structures")

# Count structures by retinal type
retinal_type_counts = {}
for pdb_id, props in retinal_properties.items():
    if props['has_retinal']:
        for ret_type in props['retinal_types']:
            retinal_type_counts[ret_type] = retinal_type_counts.get(ret_type, 0) + 1

print("\nRetinal types found:")
for ret_type, count in sorted(retinal_type_counts.items(), key=lambda x: x[1], reverse=True):
    print(f"  {ret_type}: {count} structures")

# Get statistics on retinal atom counts
if structures_with_retinal:
    atom_counts = [props['retinal_atom_count'] for pdb_id, props in retinal_properties.items() 
                  if props['has_retinal']]
    
    print(f"\nRetinal atom count statistics:")
    print(f"  Min: {min(atom_counts)}")
    print(f"  Max: {max(atom_counts)}")
    print(f"  Mean: {np.mean(atom_counts):.1f}")
    print(f"  Median: {np.median(atom_counts)}")
    
    # Get distribution of atom counts
    from collections import Counter
    atom_count_dist = Counter(atom_counts)
    
    print("\nDistribution of retinal atom counts:")
    for count, freq in sorted(atom_count_dist.items()):
        print(f"  {count} atoms: {freq} structures")
    
    # Get most common atoms
    all_atoms = []
    for pdb_id, props in retinal_properties.items():
        if props['has_retinal']:
            all_atoms.extend(props['unique_retinal_atoms'])
    
    if all_atoms:
        from collections import Counter
        atom_counter = Counter(all_atoms)
        
        print("\nMost common retinal atoms:")
        for atom, count in atom_counter.most_common(15):
            print(f"  {atom}: {count} structures")
    
# Find structures with potential issues
print("\nStructures with potential retinal issues:")
for pdb_id, props in retinal_properties.items():
    # Missing retinal
    if not props['has_retinal']:
        continue  # Too many to list individually
    
    # Low atom count (potential fragmented retinal)
    elif props['retinal_atom_count'] < 20:  # Retinal should have 20+ atoms
        print(f"  {pdb_id}: Low retinal atom count ({props['retinal_atom_count']})")
    
    # Multiple retinal molecules
    elif len(props['retinal_residues']) > 1:
        print(f"  {pdb_id}: Multiple retinal residues found: {props['retinal_residues']}")

# Count structures missing retinal
missing_retinal = len(retinal_properties) - len(structures_with_retinal)
print(f"\n{missing_retinal} structures have no retinal detected")
```

```python
# Demo of LYR processing implementation in notebook context
def demo_lyr_processing(data):
    """
    Demonstrate LYR processing on the loaded data
    
    Args:
        data (dict): Data dictionary from the workflow
        
    Returns:
        dict: Modified data with LYR residues processed
    """
    # First check which structures might have LYR residues
    structures_with_lyr = []
    
    # Check raw structures for LYR residues
    if 'raw_structures' in data:
        for pdb_id, struct_data in data['raw_structures'].items():
            if 'df' in struct_data and isinstance(struct_data['df'], pd.DataFrame):
                df = struct_data['df']
                if 'res_name3l' in df.columns and 'LYR' in df['res_name3l'].values:
                    structures_with_lyr.append(pdb_id)
    
    if not structures_with_lyr:
        print("No structures with LYR residues found in raw data.")
        return data
    
    print(f"Found {len(structures_with_lyr)} structures with LYR residues: {', '.join(structures_with_lyr)}")
    
    # We will create a modified copy of the data
    modified_data = data.copy()
    
    # Function to convert LYR to LYS+RET (using the implementation from the LYR module)
    def convert_lyr_to_lys_ret(df, retinal_name='RET'):
        if df is None or df.empty or 'res_name3l' not in df.columns or 'LYR' not in df['res_name3l'].values:
            return df
        
        # Make a copy
        result_df = df.copy()
        
        # Identify LYR residues
        lyr_mask = result_df['res_name3l'] == 'LYR'
        lyr_data = result_df[lyr_mask].copy()
        
        if lyr_data.empty:
            return result_df
        
        print(f"Processing {len(lyr_data)} LYR atoms")
        
        # Find unique LYR residue IDs
        if 'auth_seq_id' in lyr_data.columns:
            lyr_residue_ids = lyr_data['auth_seq_id'].unique()
            print(f"LYR residues at positions: {lyr_residue_ids}")
        
        # Split LYR atoms into lysine (LYS) and retinal (RET) parts
        # Lysine atoms: N, CA, C, O, CB, CG, CD, CE, NZ (protein backbone/sidechain)
        lys_atom_pattern = re.compile(r'^(N|CA|C|O|CB|CG|CD|CE|NZ)$')
        
        # Create masks for lysine and retinal atoms
        if 'res_atom_name' in lyr_data.columns:
            lys_mask = lyr_data['res_atom_name'].str.match(lys_atom_pattern)
            
            # Lysine atoms (backbone + sidechain)
            lys_atoms = lyr_data[lys_mask].copy()
            # Retinal atoms (everything else)
            ret_atoms = lyr_data[~lys_mask].copy()
            
            print(f"Split into {len(lys_atoms)} lysine atoms and {len(ret_atoms)} retinal atoms")
            
            # Update lysine atoms
            lys_atoms['res_name3l'] = 'LYS'
            if 'group' in lys_atoms.columns:
                lys_atoms['group'] = 'ATOM'
            
            # Update retinal atoms
            ret_atoms['res_name3l'] = retinal_name
            if 'group' in ret_atoms.columns:
                ret_atoms['group'] = 'HETATM'
            
            # Assign new residue ID for retinal
            if 'auth_seq_id' in result_df.columns:
                try:
                    # Extract numeric part of residue IDs and find maximum
                    max_resid = pd.to_numeric(result_df['auth_seq_id'], errors='coerce').max()
                    new_ret_id = int(max_resid + 1000) if not pd.isna(max_resid) else 1000
                    print(f"Assigning new residue ID {new_ret_id} to retinal atoms")
                    ret_atoms['auth_seq_id'] = new_ret_id
                    
                    # If label_seq_id is present, update it too
                    if 'label_seq_id' in ret_atoms.columns:
                        ret_atoms['label_seq_id'] = new_ret_id
                except Exception as e:
                    print(f"Error assigning new residue ID: {e}")
                    import traceback
                    traceback.print_exc()
            
            # Remove original LYR residues and add new LYS and RET residues
            result_df = pd.concat([result_df[~lyr_mask], lys_atoms, ret_atoms], ignore_index=True)
            
        return result_df
    
    # Process the raw structures
    if 'raw_structures' in modified_data:
        for pdb_id in structures_with_lyr:
            if pdb_id in modified_data['raw_structures']:
                struct_data = modified_data['raw_structures'][pdb_id]
                if 'df' in struct_data and isinstance(struct_data['df'], pd.DataFrame):
                    print(f"\nProcessing LYR residues in structure {pdb_id}:")
                    modified_data['raw_structures'][pdb_id]['df'] = convert_lyr_to_lys_ret(struct_data['df'])
    
    # Verify the changes
    lyr_after_processing = []
    ret_after_processing = []
    
    for pdb_id in structures_with_lyr:
        if pdb_id in modified_data['raw_structures']:
            df = modified_data['raw_structures'][pdb_id].get('df')
            if df is not None and 'res_name3l' in df.columns:
                if 'LYR' in df['res_name3l'].values:
                    lyr_after_processing.append(pdb_id)
                if 'RET' in df['res_name3l'].values:
                    ret_after_processing.append(pdb_id)
    
    print("\n=== PROCESSING RESULTS ===")
    print(f"Structures still with LYR after processing: {len(lyr_after_processing)}")
    print(f"Structures with RET after processing: {len(ret_after_processing)}")
    
    for pdb_id in ret_after_processing:
        if pdb_id in modified_data['raw_structures']:
            df = modified_data['raw_structures'][pdb_id].get('df')
            if df is not None and 'res_name3l' in df.columns:
                ret_count = df[df['res_name3l'] == 'RET'].shape[0]
                print(f"  {pdb_id}: {ret_count} RET atoms")
    
    return modified_data

# Run the LYR processing demo if we have data
if 'raw_structures' in data:
    print("=== LYR PROCESSING DEMO ===")
    print("This demonstrates how LYR processing would work on the loaded data.")
    
    # Run the demo with a copy of the data
    modified_data = demo_lyr_processing(data)
    
    # See if this fixed any structures missing retinal
    if 'processed_structures' in data and 'processed_structures' in modified_data:
        orig_structures = set(data['processed_structures'].keys())
        mod_structures = set(modified_data['processed_structures'].keys())
        
        # Check if we have different structures
        if len(orig_structures) != len(mod_structures):
            print(f"\nOriginal structure count: {len(orig_structures)}")
            print(f"Modified structure count: {len(mod_structures)}")
            
            # Show differences
            added = mod_structures - orig_structures
            if added:
                print(f"Added structures: {added}")
            
            removed = orig_structures - mod_structures
            if removed:
                print(f"Removed structures: {removed}")
    
    print("\nNote: To properly implement this fix, create the LYR processing module")
    print("in the project source files and modify the data loading process to")
    print("handle LYR residues correctly.")
else:
    print("No raw structure data available to demonstrate LYR processing.")
```
